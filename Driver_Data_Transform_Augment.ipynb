```python
import numpy as np
import os
from imageio import imread
import cv2 as cv
from cv2 import resize

np.random.seed(30)
import random as rn
rn.seed(30)
import PIL.Image
import torch
from torchvision import transforms

def imshow_c1_t6(new_folder_name,img,transform1,transform2,transform3,transform4,transform5,transform6):
    """helper function to show data augmentation
    :param img: path of the image
    :param transform: data augmentation technique to apply"""
    
    orig_img = PIL.Image.open(img)
    image_name = img.split('.')[0].split('/')[-1]
    
    img1 = transform1(orig_img)
    img1.save(new_folder_name+'_cam1_1'+'/'+image_name+'_t1'+'.jpg','JPEG')
    
    img2 = transform2(orig_img)
    img2.save(new_folder_name+'_cam1_2'+'/'+image_name+'_t2'+'.jpg','JPEG')
    
    img3 = transform3(orig_img)
    img3.save(new_folder_name+'_cam1_3'+'/'+image_name+'_t3'+'.jpg','JPEG')
    
    img4 = cv.cvtColor(cv.imread(new_folder_name+'_cam1_1'+'/'+image_name+'_t1'+'.jpg'), cv.COLOR_BGR2RGB)
    cv.imwrite(new_folder_name+'_cam1_4'+'/'+image_name+'_t4'+'.jpg', img4)    
    
    img5 = transform5(orig_img)
    img5.save(new_folder_name+'_cam1_5'+'/'+image_name+'_t5'+'.jpg','JPEG')
    
    img6 = transform6(orig_img)
    img6.save(new_folder_name+'_cam1_6'+'/'+image_name+'_t6'+'.jpg','JPEG')

def data_augmentor(source_path, folder_list):
    print( 'Source path = ', source_path)

    img_idx = [x for x in range(0,20) ] #create a list of image numbers you want to use for a particular video
    
    loader_transform_1 = transforms.Compose([transforms.Resize((240, 320)),transforms.CenterCrop(240)])    
    loader_transform_2 = transforms.Compose([transforms.Resize((240, 240))])
    loader_transform_3 = transforms.Compose([transforms.CenterCrop((360, 360)),transforms.Resize((240, 240))])
    loader_transform_4 = transforms.Compose([transforms.Resize((240, 320)),transforms.CenterCrop(240),transforms.RandomHorizontalFlip(p=1)])    
    loader_transform_5 = transforms.Compose([transforms.Resize((240, 320)),transforms.CenterCrop(240),transforms.GaussianBlur(23, (0.1, 2.0))])
    loader_transform_6 = transforms.Compose([transforms.Resize((240, 320)),transforms.CenterCrop(240),transforms.Grayscale(3)])
    
    t = np.random.permutation(folder_list)
    num_folders = len(folder_list) 
    for folder in range(num_folders): # we iterate over the number of batches
        
        folder_name = t[folder].strip().split(';')[0]
        imgs = os.listdir(source_path+'/'+folder_name) # read all the images in the folder
        new_folder_name = 'Project_Data/train'+'_cam_1_2_mv2_t6/'+ folder_name
        os.makedirs(new_folder_name+'_cam1_1')
        os.makedirs(new_folder_name+'_cam1_2')
        os.makedirs(new_folder_name+'_cam1_3')
        os.makedirs(new_folder_name+'_cam1_4')
        os.makedirs(new_folder_name+'_cam1_5')
        os.makedirs(new_folder_name+'_cam1_6')
        
        outF = open("./Project_Data/train_cam_1_2_mv2_t6.csv", "a")
        outF.write(folder_name+'_cam1_1'+';'+t[folder].strip().split(';')[1])
        outF.write("\n")
        outF.write(folder_name+'_cam1_2'+';'+t[folder].strip().split(';')[1])
        outF.write("\n")
        outF.write(folder_name+'_cam1_3'+';'+t[folder].strip().split(';')[1])
        outF.write("\n")
        outF.write(folder_name+'_cam1_4'+';'+t[folder].strip().split(';')[1])
        outF.write("\n")
        outF.write(folder_name+'_cam1_5'+';'+t[folder].strip().split(';')[1])
        outF.write("\n")
        outF.write(folder_name+'_cam1_6'+';'+t[folder].strip().split(';')[1])
        outF.write("\n")
        outF.close()
        
        for idx,item in enumerate(img_idx): #  Iterate iver the frames/images of a folder to read them in
            imshow_c1_t6(new_folder_name,source_path+'/'+folder_name+'/'+imgs[item],loader_transform_1,loader_transform_2,loader_transform_3,loader_transform_4,loader_transform_5,loader_transform_6)

train_path = 'Project_Data/train_cam1'

train_doc = np.random.permutation(open('./Project_Data/train_cam1.csv').readlines())

data_augmentor(train_path, train_doc)
```
